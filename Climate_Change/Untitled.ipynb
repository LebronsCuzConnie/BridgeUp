```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("Salinity_info_38N_68W.csv")
```

```python
df.head()
```

```python
depth = df["Depth "]
tempC = df["Temperature (Celsius)"]
salinity = df["Salinity (PSU)"]
density = df["Density (kg/m3)"]
```

```python
plt.plot(depth, salinity, color="magenta", label="Salinity")
plt.xlabel("Depth")
plt.legend(loc=9)
plt.ylabel("Salinity")
plt.twinx()

plt.title("Salinity in relation to the temperature the Ocean")
plt.plot(depth, tempC, color="blue", label="Temperature(C)")
plt.legend()


plt.ylabel("Temperature")
plt.show()


```

```python
plt.plot(depth, salinity, color="magenta", label="Salinity")
plt.xlabel("Depth")
plt.legend(loc=9)
plt.ylabel("Salinity")
plt.twinx()

plt.title("Salinity in relation to the Density of the Ocean")
plt.plot(depth, density, color="blue", label="Density (kg/m3)")
plt.legend()


plt.ylabel("Temperature")
plt.show()

```
