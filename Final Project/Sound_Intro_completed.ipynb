# Creating sound with Python

First, import numpy and matplotlib.pyplot:

```python
import numpy as np
import matplotlib.pyplot as plt
```

We will also need to import the LibROSA library for music and audio analysis:

```python
import librosa # notice that there is no alias so we will refer to the entire library in our code
```

## First, let's create a sine wave

Let's define pi, the number of points per cycle, and the number of cycles:

```python
pi  = np.pi # pi (notice that we will be using a lot of numpy functions)
npts_cycle = 40 # number of points per cycle, or samples per cycle
n_cycles = 3 # number of cycles
```

We can use those variables to calculate the x and y values for our graph:

```python
x = np.linspace(0,2*pi*n_cycles,npts_cycle*n_cycles)
y = np.sin(x)
```

Now let's plot our sine wave. 
Use plt.plot(). Label the x axis "radians" and the y axis "wave amplitude"

```python
plt.plot(x,y) 
plt.xlabel('radians')
plt.ylabel('amplitude of wave')
plt.show()
```

## Adding time and frequency
Our sine wave is missing something important for auditory information: **frequency**. 

Without frequency we do not know how fast the sine wave is moving.

Let's add some more cycles to our wave. Redefine n_cycles as 440:

```python
n_cycles = 440
```

Now we can calculate frequency (number of cycles per second):

```python
p = 1/n_cycles # p stands for period and is calculated in seconds
f = 1/p # f stands for frequency

print("frequency (# cycles/second) = ", f)
```

We will not be plotting all the infinite points on our wave. Let's calculate our sampling frequency (number of samples per second):

```python
dt = 1/f/npts_cycle 
fs = int(1/dt) # fs stands for sampling frequency (note, not the same as f, which is our wave frequency)

print("sampling frequency (# samples/second) = ", fs)
```

Now we can plot our sine wave with time on the x axis to see how far the wave moves in one second:

```python
x = np.arange(0,(p*n_cycles),dt)

a = np.linspace(0,2*pi*n_cycles,npts_cycle*n_cycles)
y = np.sin(a)
```

Write the plotting code here:

```python
plt.plot(x,y)
plt.xlabel("time (sec)")
plt.ylabel("amplitude of wave")
plt.show()
```

Our frequency is so high that we can't even see the individual cycles at this scale. 

Try replotting with an x limit of 3 times the period (p*3):

```python
plt.plot(x,y)
plt.xlabel("time (sec)")
plt.ylabel("amplitude of wave")
plt.xlim(0, p*3)
plt.show()
```

Notice that we get three repeats of our wave in 0.006 seconds!

## Making a sound file
First let's create the output file:

```python
outfile = "sample_sound.wav"
```

We will use the librosa output.write_wav() function to create our sound file:

```python
a_scale = 0.2 # this scales the amplitude
librosa.output.write_wav(outfile, y*a_scale, fs, norm=False) # we are using our y data and sampling frequency

# norm=False means the signal will not be normalized (default is True). 
```
